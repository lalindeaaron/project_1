## Load Datasets from csv

```python
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')

sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')

products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')

transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
users.head()
```

```python
sessions.head()
```

```python
transactions.head()
```

```python
#left join with matching user id
left, right = transactions, users
result = left.merge(right, on=['UserID'], how='left')
```

```python
result.head(5)
```

```python
# Which transactions have a UserID not in users?

transactions[~transactions.UserID.isin(users['UserID'])]
```

```python
#Join users to transactions, keep only rows from transactions and users that match via UserID (inner join)
result = left.merge(right, on=['UserID'], how='inner')
```

```python
result.head(5)
```

```python
#Determine which sessions occurred on the same day each user registered
pd.merge(users,sessions,left_on=['UserID','Registered'], right_on=['UserID','SessionDate'],how='inner')
```

```python
#Build a dataset with every possible (UserID, ProductID) pair (cross join)
userdf = pd.DataFrame({'key': np.repeat(1, users.shape[0]), 'UserID': users.UserID})
productdf=pd.DataFrame({'key':np.repeat(1,products.shape[0]),'ProductID':products.ProductID})
pd.merge(userdf,productdf,on='key')[['UserID','ProductID']]
```

```python
# Determine how much quantity of each product was purchased by each user
purchased=pd.merge(userdf,productdf,on='key')[['UserID','ProductID']]

pd.merge(purchased, transactions, how='left', on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(lambda x: pd.Series(dict(Quantity=x.Quantity.sum())))
```

```python
#For each user, get each possible pair of pair transactions (TransactionID1,TransacationID2)
pd.merge(transactions,transactions,on='UserID')
```

```python
#Join each users to his/her first occuring transaction in the transactions table??
pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')
```
